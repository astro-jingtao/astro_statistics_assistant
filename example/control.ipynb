```python
from asa.control import control_1d
from asa.plot_methods import plot_hist
import numpy as np
```

```python
x_A = np.random.uniform(-1, 1, 1000)
x_B = np.random.normal(size=1000)
```

```python
plot_hist(x_A,
          range=(-3, 3),
          density=True,
          bins=31,
          ecolor='blue',
          color='blue',
          alpha=0.5)
plot_hist(x_B,
          range=(-3, 3),
          density=True,
          bins=31,
          ecolor='red',
          color='red',
          alpha=0.5)
```

```python
A_index, B_index, _, _ = control_1d(x_A, x_B, bins=31, _range=(-3, 3))
```

```python
plot_hist(x_A[A_index],
          range=(-3, 3),
          density=True,
          bins=31,
          ecolor='blue',
          color='blue',
          alpha=0.5)
plot_hist(x_B[B_index],
          range=(-3, 3),
          density=True,
          bins=31,
          ecolor='red',
          color='red',
          alpha=0.5)
```
