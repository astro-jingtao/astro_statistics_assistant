```python
from asa.linear_model import get_OLS, get_WLS, get_ODR
from asa.plot_methods import plot_line
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('science')
plt.rcParams.update({'font.size': 15})
```

## x uniformly distributed

```python
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(0, 0.1, 100)

y_err = np.abs(y) / 3
y_obs = y + np.random.normal(0, y_err, 100)

x_err = np.abs(x) / 3
x_obs = x + np.random.normal(0, x_err, 100)
```

```python
res_OLS = get_OLS(x_obs, y_obs)
res_WLS = get_WLS(x_obs, y_obs, y_err=y_err)
res_ODR = get_ODR(x_obs, y_obs, x_err=x_err, y_err=y_err)
```

```python
plt.figure(figsize=(8, 8), dpi=400)
plt.scatter(x_obs, y_obs)
plot_line(k=2, b=1, color='black', label='Real')
plot_line(k=res_OLS['k'][0], b=res_OLS['b'][0], color='red', label='OLS')
plot_line(k=res_WLS['k'][0], b=res_WLS['b'][0], color='green', label='WLS')
plot_line(k=res_ODR['k'][0], b=res_ODR['b'][0], color='blue', label='ODR')

plt.legend()
```
